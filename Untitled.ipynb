# Penalty Aggregate

```python
def get_penalty_aggregate(df):
    penalty_aggregate = {}
    for index, row in df.iterrows():
        year = row['game_date'][:4]
        if row['penalty_type'] in penalty_aggregate:
            if year in penalty_aggregate[row['penalty_type']]:
                penalty_aggregate[row['penalty_type']][year] += 1
            else:
                penalty_aggregate[row['penalty_type']][year] = 1
        else:
            penalty_aggregate[row['penalty_type']] = {}
            penalty_aggregate[row['penalty_type']][year] = 1
    return penalty_aggregate
penalty_aggregate = get_penalty_aggregate(xl)
```

# Kickoff Offense/Defense

```python
kickoff_off_def_split={'2009':{'offense':0,'defense':0},'2010':{'offense':0,'defense':0},'2011':{'offense':0,'defense':0},
                       '2012':{'offense':0,'defense':0},'2013':{'offense':0,'defense':0},'2014':{'offense':0,'defense':0},
                       '2015':{'offense':0,'defense':0},'2016':{'offense':0,'defense':0},'2017':{'offense':0,'defense':0},
                       '2018':{'offense':0,'defense':0}}
for index, row in xl.iterrows():
    if row['play_type'] != 'kickoff':
        continue
    penalty_team = row['penalty_team']
    if penalty_team == row['posteam']:
        offense_penalty = True
    elif penalty_team == row['defteam']:
        offense_penalty = False
    else:
        print(row['penalty_type'])
        continue
        
    year = row['game_date'][:4]
    if offense_penalty:
        kickoff_off_def_split[year]['offense'] += 1
    else:
        kickoff_off_def_split[year]['defense'] += 1
kickoff_off_def_split    
```

```python
df = pd.DataFrame(kickoff_off_def_split).T
df.index.name = 'Year'
df = df.reset_index()

tempo = df['offense'].copy()
tempd = df['defense'].copy()

df["offense"] = tempo/(tempo+tempd)
df["defense"] = tempd/(tempo+tempd)
df_total = df["offense"] + df["defense"]
df_rel = df[df.columns[1:]].div(df_total, 0)*100

df.plot(
  x = 'Year', 
  kind = 'barh', 
  stacked = True, 
  title = 'Percentage Stacked Bar Graph', 
  mark_right = True)
  
for n in df_rel:
    for i, (cs, ab, pc) in enumerate(zip(df.iloc[:, 1:].cumsum(1)[n], 
                                         df[n], df_rel[n])):
        plt.text(cs - ab / 2, i, str(np.round(pc, 1)) + '%', 
                 va = 'center', ha = 'center')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
```

In 2011 and 2016, we see increases in the percentage of offense penalties, which are the years in which kickoff position rules changed. The rules were designed to target/incentivize offenses to take more touchbacks, but the number of kickoff penalties have gone up, with the offense having a bigger share of those penalties. This further supports the idea that the rules are not effective.

# Kickoff - Should they have returned it?

```python
xl['shifted_yardline_100'] = xl['yardline_100'].shift(-1).fillna(0)
kickoff_decision={'2009':{'right':0,'wrong':0},'2010':{'right':0,'wrong':0},'2011':{'right':0,'wrong':0},
                       '2012':{'right':0,'wrong':0},'2013':{'right':0,'wrong':0},'2014':{'right':0,'wrong':0},
                       '2015':{'right':0,'wrong':0},'2016':{'right':0,'wrong':0},'2017':{'right':0,'wrong':0},
                       '2018':{'right':0,'wrong':0}}
total_kickoff_return={'2009':0,'2010':0,'2011':0,'2012':0,'2013':0,'2014':0,'2015':0,'2016':0,'2017':0,'2018':0}
for index, row in xl.iterrows():
    # check if kickoff
    if row['play_type'] != "kickoff":
        continue
        
    # check if into endzone
    if not (row['kickoff_in_endzone'] == 1):
        continue

    # check yardage
    yardline_raw = row['shifted_yardline_100']
    yardline = 100-yardline_raw
    year = row['game_date'][:4]
    if int(year) < 2016:
        if yardline < 20:
            kickoff_decision[year]['wrong']+=1
        else:
            kickoff_decision[year]['right']+=1
    else:
        if yardline < 25:
            kickoff_decision[year]['wrong']+=1
        else:
            kickoff_decision[year]['right']+=1
    total_kickoff_return[year]+=1
```

```python
df = pd.DataFrame(kickoff_decision).T
df.index.name = 'Year'
df = df.reset_index()

tempo = df['right'].copy()
tempd = df['wrong'].copy()

df["right"] = tempo/(tempo+tempd)
df["wrong"] = tempd/(tempo+tempd)
df_total = df["right"] + df["wrong"]
df_rel = df[df.columns[1:]].div(df_total, 0)*100

df.plot(
  x = 'Year', 
  kind = 'barh', 
  stacked = True, 
  title = 'Percentage Stacked Bar Graph', 
  mark_right = True)
  
for n in df_rel:
    for i, (cs, ab, pc) in enumerate(zip(df.iloc[:, 1:].cumsum(1)[n], 
                                         df[n], df_rel[n])):
        plt.text(cs - ab / 2, i, str(np.round(pc, 1)) + '%', 
                 va = 'center', ha = 'center')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
```

```python
plt.figure()

x = total_kickoff_return.keys()
y = total_kickoff_return.values()
plt.bar(x, y, label='Kickoffs Returned')
plt.xlabel("year")
plt.ylabel("count")
plt.legend()
plt.title("Kickoffs Returned Over Years")
```

We see that in both 2011 and 2016, there is a drop in 'right' percentage, which means that it is rarely the right move to return the ball, yet they are still doing it. Especially in 2011, when even more kickoffs were returned, indicating that the rules had the opposite effect than intended. There was a significant drop in kickoffs returned in 2016

# Kickoff return strategy by team

```python
import copy

xl['shifted_yardline_100'] = xl['yardline_100'].shift(-1).fillna(0)
kickoff_decision_temp={'2009':{'right':0,'wrong':0},'2010':{'right':0,'wrong':0},'2011':{'right':0,'wrong':0},
                       '2012':{'right':0,'wrong':0},'2013':{'right':0,'wrong':0},'2014':{'right':0,'wrong':0},
                       '2015':{'right':0,'wrong':0},'2016':{'right':0,'wrong':0},'2017':{'right':0,'wrong':0},
                       '2018':{'right':0,'wrong':0}}
total_kickoff_return_temp={'2009':0,'2010':0,'2011':0,'2012':0,'2013':0,'2014':0,'2015':0,'2016':0,'2017':0,'2018':0}
kickoff_decision_team={}
total_kickoff_return_team={}
total_kickoff_touchback_return_team={}
for index, row in xl.iterrows():
    # check if kickoff
    if row['play_type'] != "kickoff":
        continue
        
    # check if touchback
    if "touchback" in row['desc'].lower():
        team_name = row['posteam']
        if team_name not in total_kickoff_touchback_return_team:
            total_kickoff_touchback_return_team[team_name]=copy.deepcopy(total_kickoff_return_temp)
        year = row['game_date'][:4]
        total_kickoff_touchback_return_team[team_name][year]+=1
        
    # check if into endzone
    if not (row['kickoff_in_endzone'] == 1):
        continue

    # get team name
    team_name = row['posteam']
    if team_name not in kickoff_decision_team:
        kickoff_decision_team[team_name]=copy.deepcopy(kickoff_decision_temp)
    if team_name not in total_kickoff_return_team:
        total_kickoff_return_team[team_name]=copy.deepcopy(total_kickoff_return_temp)
    
    # check yardage
    yardline_raw = row['shifted_yardline_100']
    yardline = 100-yardline_raw
    year = row['game_date'][:4]
    if int(year) < 2016:
        if yardline < 20:
            kickoff_decision_team[team_name][year]['wrong']+=1
        else:
            kickoff_decision_team[team_name][year]['right']+=1
    else:
        if yardline < 25:
            kickoff_decision_team[team_name][year]['wrong']+=1
        else:
            kickoff_decision_team[team_name][year]['right']+=1
    total_kickoff_return_team[team_name][year]+=1
```

```python
team_totals = {}
for team in kickoff_decision_team:
    data = kickoff_decision_team[team]
    temp_right = 0
    temp_wrong = 0
    for year in data:
        temp_right += data[year]['right'] 
        temp_wrong += data[year]['wrong']
    team_totals[team] = {'right':temp_right,'wrong':temp_wrong}
```

```python
# merge STL and LA, SD and LAC, JAC and JAX
for year in kickoff_decision_team['LA']:
    kickoff_decision_team['LA'][year]={key:kickoff_decision_team['LA'][year].get(key,0) + kickoff_decision_team['STL'][year].get(key,0)
     for key in kickoff_decision_team['LA'][year].keys() | kickoff_decision_team['STL'][year].keys()}
total_kickoff_return_team['LA']={key:total_kickoff_return_team['LA'].get(key,0) + total_kickoff_return_team['STL'].get(key,0)
 for key in total_kickoff_return_team['LA'].keys() | total_kickoff_return_team['STL'].keys()}
total_kickoff_touchback_return_team['LA']={key:total_kickoff_touchback_return_team['LA'].get(key,0) + total_kickoff_touchback_return_team['STL'].get(key,0)
 for key in total_kickoff_touchback_return_team['LA'].keys() | total_kickoff_touchback_return_team['STL'].keys()}
    
for year in kickoff_decision_team['LAC']:
    kickoff_decision_team['LAC'][year]={key:kickoff_decision_team['LAC'][year].get(key,0) + kickoff_decision_team['SD'][year].get(key,0)
     for key in kickoff_decision_team['LAC'][year].keys() | kickoff_decision_team['SD'][year].keys()}
total_kickoff_return_team['LAC']={key:total_kickoff_return_team['LAC'].get(key,0) + total_kickoff_return_team['SD'].get(key,0)
 for key in total_kickoff_return_team['LAC'].keys() | total_kickoff_return_team['SD'].keys()}
total_kickoff_touchback_return_team['LAC']={key:total_kickoff_touchback_return_team['LAC'].get(key,0) + total_kickoff_touchback_return_team['SD'].get(key,0)
 for key in total_kickoff_touchback_return_team['LAC'].keys() | total_kickoff_touchback_return_team['SD'].keys()}
    
for year in kickoff_decision_team['JAX']:
    kickoff_decision_team['JAX'][year]={key:kickoff_decision_team['JAX'][year].get(key,0) + kickoff_decision_team['JAC'][year].get(key,0)
     for key in kickoff_decision_team['JAX'][year].keys() | kickoff_decision_team['JAC'][year].keys()}
total_kickoff_return_team['JAX']={key:total_kickoff_return_team['JAX'].get(key,0) + total_kickoff_return_team['JAC'].get(key,0)
 for key in total_kickoff_return_team['JAX'].keys() | total_kickoff_return_team['JAC'].keys()}
total_kickoff_touchback_return_team['JAX']={key:total_kickoff_touchback_return_team['JAX'].get(key,0) + total_kickoff_touchback_return_team['JAC'].get(key,0)
 for key in total_kickoff_touchback_return_team['JAX'].keys() | total_kickoff_touchback_return_team['JAC'].keys()}
    
print(kickoff_decision_team['LA'])
```
